# Prep before training

```python
x = datos_clima_promedios.drop(["Temp_Air_Mean"], axis=1)
x = x.loc[2020:2023]
x
```

```python
y = producciones[["Cantidad", "Variedad"]]
y
```

```python
_x = x.drop(["Relat_Hum_Min", "Wind_Speed_Min"], axis=1)
x_tommy = pd.concat([_x]*2, ignore_index=True)
x_tommy
```

```python
_x = x.drop(["Temp_Air_Max", "Wind_Speed_Max"], axis=1)
x_ataulfo = pd.concat([_x]*1, ignore_index=True)
x_ataulfo
```

```python
y_tommy = y[y["Variedad"] == "Tommy Atkins"]["Cantidad"]#.values
y_tommy
```

```python
y_ataulfo = y[y["Variedad"] == "Ataulfo"]["Cantidad"]#.values
y_ataulfo
```

# Model Training

## Decision trees

```python
# Create Decision Tree classifer object
linear_regressor_tommy = LinearRegression().fit(x_tommy, y_tommy)
linear_regressor_tommy
```

```python
# Create Decision Tree classifer object
linear_regressor_ataulfo = LinearRegression().fit(x_ataulfo, y_ataulfo)
linear_regressor_ataulfo
```

## Random Forest

```python
from sklearn.linear_model import LinearRegression
```

```python
# Create Decision Tree classifer object
linear_regressor_tommy = LinearRegression().fit(x_tommy, y_tommy)
linear_regressor_tommy
```

```python
# Create Decision Tree classifer object
forest_regressor_ataulfo = LinearRegression().fit(x_ataulfo, y_ataulfo)
forest_regressor_ataulfo
```

```python
from sklearn.model_selection import cross_val_score, LeaveOneOut

def leave_one_out_validate(model, x, y):
    # Perform LOOCV using cross_val_score
    # Pass in the regressor, feature matrix X, target variable y, and the LOO iterator
    scores = cross_val_score(forest_regressor_tommy, x_tommy, y_tommy, cv=LeaveOneOut(), scoring='neg_mean_squared_error')

    # Convert the scores to positive values and calculate the mean
    mean_score = -scores.mean()

    # Print the mean squared error obtained from LOOCV
    return mean_score
```

```python
print("Mean Squared Error (LOOCV):", leave_one_out_validate(linear_regressor_tommy, x_tommy, y_tommy))
print("Mean Squared Error (LOOCV):", leave_one_out_validate(linear_regressor_ataulfo, x_ataulfo, y_ataulfo))
```
