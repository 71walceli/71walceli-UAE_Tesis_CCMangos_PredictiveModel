# Transformación de datos

## Identificación de datos inválidos

```python
variables_ambientales = [
    "Precipitation",
    "Temp_Air_Mean",
    "Temp_Air_Min",
    "Temp_Air_Max",
    "Dew_Temp_Mean",
    "Dew_Temp_Max",
    "Dew_Temp_Min",
    "Relat_Hum_Mean",
    "Relat_Hum_Min",
    "Relat_Hum_Max",
    "Wind_Speed_Mean",
    "Wind_Speed_Min",
    "Wind_Speed_Max",
    "Atmospheric_Pressure_Max",
    "Atmospheric_Pressure_Min",
]
```

```python
años = datos_clima["Date"].map(lambda v: v[:4])
años
```

```python
porcentaje_nulos_anuales = datos_clima.isnull().groupby(años).aggregate(lambda series: sum(series)/len(series))
display_full(porcentaje_nulos_anuales)
```

```python
porcentaje_nulos_anuales.loc["Total"] = \
    porcentaje_nulos_anuales.sum(axis=0, numeric_only=True)/porcentaje_nulos_anuales.count(axis=0, numeric_only=True)
porcentaje_nulos_anuales.loc[:,"Total"] = \
    porcentaje_nulos_anuales.sum(axis=1, numeric_only=True)/porcentaje_nulos_anuales.count(axis=1, numeric_only=True)
display_full(porcentaje_nulos_anuales)
```

```python
display_full(porcentaje_nulos_anuales.sort_values("Total", ascending=False))
```

```python
display_full(porcentaje_nulos_anuales.transpose().sort_values("Total", ascending=False))
```

## Limpieza y eliminación

### Variables

```python
variables_incompletas = [
    "Precipitation",
    "Dew_Temp_Mean",
    "Dew_Temp_Max",
    "Dew_Temp_Min",
    "Atmospheric_Pressure_Max",
    "Atmospheric_Pressure_Min",
]
```

```python
datos_clima = datos_clima[["Date",*variables_ambientales]]
datos_clima = datos_clima.drop([*variables_incompletas], axis=1)
datos_clima["Date"] = pd.to_datetime(datos_clima["Date"])
datos_clima = datos_clima.set_index("Date")
datos_clima
```

```python
producciones = producciones.drop(["FechaRegistro", "Activo", "Id_Usuario", "id"], axis=1)
producciones
```

### Registros

```python
datos_clima[datos_clima.isnull().sum(axis=1) > 0]
```

```python
datos_clima = datos_clima.drop(datos_clima[datos_clima.isnull().sum(axis=1) > 0].index)
datos_clima
```

## Conversión de tipos de datos

```python
producciones = df_change_types(producciones, {
    "Cantidad": float,
    #"Id_Area": int,
    "Id_Lote": int,
})
producciones.dtypes
```

```python
datos_clima = df_change_types(datos_clima, {column: float 
    for column in {*variables_ambientales} - {*variables_incompletas}
})
datos_clima.dtypes
```

## Integración y unión

```python
producciones["Variedad"] = producciones.Id_Lote.map(areas.set_index("id")["Variedad"])
producciones
```

## Agrupamiento

```python
producciones["años"] = producciones["Fecha"].map(lambda fecha: fecha[:4]).astype(int)
producciones.index = pd.MultiIndex.from_arrays(producciones[["Id_Lote", "años"]].values.T)
producciones.drop(["años", "Id_Lote"], axis=1, inplace=True)
producciones
```

## Sumarización

```python
datos_clima.groupby(años).mean(numeric_only=True)
```

```python
datos_clima_promedios = datos_clima.groupby(años).mean(numeric_only=True)
datos_clima_promedios.index = datos_clima_promedios.index.astype(int)
datos_clima_promedios
```

```python
datos_clima_suma = datos_clima.groupby(años).sum(numeric_only=True)
datos_clima_suma.index = datos_clima_suma.index.astype(int)
datos_clima_suma
```
